```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import time

# Загрузка и подготовка данных
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Нормализация и преобразование формата
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Добавление канала (28, 28, 1) для совместимости с Conv2D
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

# Преобразование меток в one-hot encoding
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print("Форма тренировочных данных:", x_train.shape)
print("Форма тестовых данных:", x_test.shape)

# Функция для построения модели LeNet
def build_lenet_model(input_shape=(28, 28, 1), num_classes=10):
    model = keras.Sequential([
        # Первый сверточный блок
        layers.Conv2D(6, (5, 5), activation='relu', input_shape=input_shape, padding='same'),
        layers.AveragePooling2D((2, 2)),
        
        # Второй сверточный блок
        layers.Conv2D(16, (5, 5), activation='relu'),
        layers.AveragePooling2D((2, 2)),
        
        # Полносвязные слои
        layers.Flatten(),
        layers.Dense(120, activation='relu'),
        layers.Dense(84, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Легкая версия AlexNet для изображений 28x28
def build_light_alexnet_model(input_shape=(28, 28, 1), num_classes=10):
    model = keras.Sequential([
        # Первый сверточный блок
        layers.Conv2D(16, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Второй сверточный блок
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Третий сверточный блок
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Dropout(0.25),
        
        # Глобальный пулинг
        layers.GlobalAveragePooling2D(),
        
        # Полносвязные слои
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Компиляция и обучение моделей
def train_and_evaluate_model(model, x_train, y_train, x_test, y_test, model_name, epochs=10):
    print(f"\n=== Обучение {model_name} ===")
    
    # Компиляция модели
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Callback для уменьшения learning rate
    reduce_lr = keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001
    )
    
    # Ранняя остановка
    early_stopping = keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=5, restore_best_weights=True, verbose=1
    )
    
    # Обычное обучение
    history = model.fit(
        x_train, y_train,
        batch_size=128,
        epochs=epochs,
        validation_data=(x_test, y_test),
        callbacks=[reduce_lr, early_stopping],
        verbose=1
    )
    
    # Оценка модели
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"{model_name} - Точность на тестовых данных: {test_acc:.4f}")
    
    return history, test_acc

# Создание и обучение LeNet
print("=== LeNet ===")
lenet_model = build_lenet_model()
lenet_model.summary()

lenet_history, lenet_acc = train_and_evaluate_model(
    lenet_model, x_train, y_train, x_test, y_test, "LeNet", epochs=10
)

# Создание и обучение легкой AlexNet
print("\n=== AlexNet ===")
alexnet_model = build_light_alexnet_model()
alexnet_model.summary()

alexnet_history, alexnet_acc = train_and_evaluate_model(
    alexnet_model, x_train, y_train, x_test, y_test, "AlexNet", epochs=15
)

# Визуализация результатов обучения
def plot_training_history(histories, model_names):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    colors = ['blue', 'red']
    for i, (history, name) in enumerate(zip(histories, model_names)):
        color = colors[i % len(colors)]
        
        # График точности
        ax1.plot(history.history['accuracy'], label=f'{name} train', color=color, linewidth=2)
        ax1.plot(history.history['val_accuracy'], label=f'{name} val', 
                color=color, linestyle='--', alpha=0.8)
        ax1.set_title('Точность моделей', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Эпоха')
        ax1.set_ylabel('Точность')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # График потерь
        ax2.plot(history.history['loss'], label=f'{name} train', color=color, linewidth=2)
        ax2.plot(history.history['val_loss'], label=f'{name} val', 
                color=color, linestyle='--', alpha=0.8)
        ax2.set_title('Потери моделей', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Эпоха')
        ax2.set_ylabel('Потери')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Построение графиков
plot_training_history([lenet_history, alexnet_history], 
                     ['LeNet', 'AlexNet'])

# Функция для предсказания на нескольких примерах
def show_predictions(models, model_names, x_test, y_test, num_samples=5):
    # Выбираем случайные примеры
    indices = np.random.choice(len(x_test), num_samples)
    
    for model, name in zip(models, model_names):
        plt.figure(figsize=(15, 3))
        plt.suptitle(f'Предсказания {name}', fontsize=16, fontweight='bold')
        
        correct_predictions = 0
        for i, idx in enumerate(indices):
            plt.subplot(1, num_samples, i + 1)
            
            # Предсказание
            prediction = model.predict(x_test[idx:idx+1], verbose=0)
            predicted_class = np.argmax(prediction)
            true_class = np.argmax(y_test[idx])
            
            # Отображение изображения
            plt.imshow(x_test[idx].squeeze(), cmap='gray')
            color = 'green' if predicted_class == true_class else 'red'
            if predicted_class == true_class:
                correct_predictions += 1
            plt.title(f'True: {true_class}\nPred: {predicted_class}', color=color, fontweight='bold')
            plt.axis('off')
        
        print(f"{name}: {correct_predictions}/{num_samples} правильных предсказаний")
        plt.tight_layout()
        plt.show()

# Показываем предсказания для всех моделей
show_predictions([lenet_model, alexnet_model],
                ['LeNet', 'AlexNet'],
                x_test, y_test)

# Функция для подсчета параметров
def count_parameters(model):
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    non_trainable_params = sum([tf.keras.backend.count_params(w) for w in model.non_trainable_weights])
    return trainable_params, non_trainable_params

# Сводка по производительности моделей
print("\n" + "="*50)
print("СРАВНЕНИЕ МОДЕЛЕЙ")
print("="*50)
print(f"LeNet точность:    {lenet_acc:.4f}")
print(f"AlexNet точность:  {alexnet_acc:.4f}")

print("\n" + "="*50)
print("СРАВНЕНИЕ РАЗМЕРОВ МОДЕЛЕЙ")
print("="*50)

models = [lenet_model, alexnet_model]
model_names = ['LeNet', 'AlexNet']

for model, name in zip(models, model_names):
    print(f"\n{name}:")
    print(f"  Всего параметров: {model.count_params():,}")
    
    # Детальная информация
    trainable, non_trainable = count_parameters(model)
    print(f"  Обучаемые параметры: {trainable:,}")
    print(f"  Необучаемые параметры: {non_trainable:,}")

# Сравнение времени предсказания
def compare_prediction_time(models, model_names, x_test, num_samples=1000):
    print(f"\n{'='*50}")
    print("СРАВНЕНИЕ ВРЕМЕНИ ПРЕДСКАЗАНИЯ")
    print(f"{'='*50}")
    
    test_indices = np.random.choice(len(x_test), num_samples, replace=False)
    test_batch = x_test[test_indices]
    
    for model, name in zip(models, model_names):
        # Прогрев
        model.predict(test_batch[:10], verbose=0)
        
        # Измерение времени
        start_time = time.time()
        predictions = model.predict(test_batch, verbose=0)
        end_time = time.time()
        
        avg_time = (end_time - start_time) / num_samples * 1000  # мс на одно изображение
        print(f"{name}: {avg_time:.3f} мс на изображение")

# Сравниваем время предсказания
compare_prediction_time(
    [lenet_model, alexnet_model],
    ['LeNet', 'AlexNet'],
    x_test
)

# Дополнительная информация о классах Fashion MNIST
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print("\nКлассы Fashion MNIST:")
for i, name in enumerate(class_names):
    print(f"{i}: {name}")

# Финальное сравнение
print("\n" + "="*60)
print("ИТОГОВОЕ СРАВНЕНИЕ МОДЕЛЕЙ")
print("="*60)

for model, name in zip(models, model_names):
    trainable, non_trainable = count_parameters(model)
    total_params = trainable + non_trainable
    
    # Измеряем точность еще раз для уверенности
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    
    print(f"\n{name}:")
    print(f"  Точность: {test_acc:.4f}")
    print(f"  Параметры: {total_params:,}")
    print(f"  Относительная точность: {test_acc/total_params*1e6:.6f} (точность на 1M параметров)")

print(f"\nВывод: {'AlexNet показывает лучшую точность' if alexnet_acc > lenet_acc else 'LeNet показывает лучшую точность'}")
```
